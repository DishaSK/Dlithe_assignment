```python
"""If the natural number is greater than 1 and having no positive divisors other than 1 and the number itself etc.
For example: 3, 7, 11 etc are prime numbers.""
#Try to write a program using for loop, if -else to print prime numbers between 0-20.

```

```python
#using for loop and if-else
start=0
end=20
  
for i in range(start, end+1):
  if i>1:
    for j in range(2,i):
        if(i % j==0):
            break
    else:
        print(i)
```

```python
#try to use pop() and remove() functions on a given list 
#use pop()
even=[2,4,6,8,10,12,14]
even.pop(3)
print(even)
#use remove()
even.remove(14)
print(even)
    
```

Analyse the difference between above two functions and write what have you understood:In the above lines of code,pop(),remove() functions are used.pop() takes in a single arguement i,e index. Within the parenthesis of pop(), an index number is put and the element present at that particular index in a list will be removed.
Whereas in the remove() function , the element to be removed from the list is directly put up within the parenthesis of remove() and that particular element is removed directly from the list. pop() function flows along with index values in the list whereas remove() function goes directly with values to be removed.
