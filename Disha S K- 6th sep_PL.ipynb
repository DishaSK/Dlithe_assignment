````python
#execute the above code and analyse the difference between method 1 and 2
emp = {'name': 'Aman', 'age': 30}
print('Employee Name: ', emp.get('name'))
print('Employee Age: ', emp.get('age'))
# using get - method 1
print('Employee Salary: ', emp.get('salary'))
# using get - method 2
print('Employee Salary: ', emp.get('salary', 0.0))

````

Differentiate between emp.get('salary') and emp.get('salary', 0.0)


ANSWER:-----In the method 1 of the code,emp.get('salary') is used,wherein the dictionary named emp does not contain the key 'salary' and hence it gives the output as None.
In the method 2 of the code,emp.get('salary', 0.0) is used,wherein the same dictionary named emp which does not contain the key 'salary' . Instead of showing None as the output,0.0 is given as the arguement within the .get() function. Hence output turns out to be 0.0 for the key 'salary'.





Have you worked on enumerate function?
Lets see how it works...
Python enumerate() function takes any iterable as argument and returns enumerate object using which the iterable can be traversed. 

It contains **index and corresponding item** in the iterable object like list, tuple or string.


run the below code ... you will see that each item in string 's' is assigned an index by enumerate function

````python
s='hello'
for i,j in enumerate(s):
  print(i,j)
````

````python
#Make use of enumerate() to create a dictionary on days--Mon, Tue, Wed.. 
days=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
dictionary={}
for index,item in enumerate(days):
  dictionary.update({index:item})

print(dictionary)








 




````
